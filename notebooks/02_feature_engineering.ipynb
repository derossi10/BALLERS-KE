```python
df["Passing_Index"] = df["ShortPassing"] * 0.5 + df["LongPassing"] * 0.5
```

```python
df["Attack_Index"] = (
    df["Finishing"] +
    df["ShotPower"] +
    df["Positioning"]
) / 3
```

```python
df["Defensive_Index"] = (
    df["StandingTackle"] +
    df["SlidingTackle"] +
    df["Interceptions"]
) / 3
```

```python
df["Overall_FWD_Score"] = (
    df["Finishing"] * 0.4 +
    df["Attack_Index"] * 0.3 +
    df["Pace"] * 0.3
)
```
