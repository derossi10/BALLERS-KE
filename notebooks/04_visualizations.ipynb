```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../data/processed/players_features.csv")

# Check structure
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Performance_Score"], bins=30, kde=True)

plt.title("Distribution of Player Performance Scores")
plt.xlabel("Performance Score")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="PositionGroup", data=df)

plt.title("Number of Players by Position")
plt.xlabel("Position Group")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="PositionGroup", y="Performance_Score", data=df)

plt.title("Performance Score by Position")
plt.xlabel("Position")
plt.ylabel("Performance Score")

plt.show()
```

```python
plt.figure(figsize=(12,8))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap="coolwarm", center=0)

plt.title("Feature Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="Passing_Index",
    y="Performance_Score",
    hue="PositionGroup",
    data=df
)

plt.title("Passing Index vs Performance Score")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="Performance_Tier", data=df)

plt.title("Player Performance Tier Distribution")

plt.show()
```
